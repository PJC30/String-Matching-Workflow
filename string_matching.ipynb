```python
import pandas as pd
import glob
import os
import operator
from itertools import combinations


def foo(it):
    it = iter(it)
    f = it.__next__()
    for s in it:
        yield f, s
        f = s

file = open("/home/astegic/python_workspace/x/input.csv")
f = len(file.readlines()) - 1

n = 0
o = 1

while n < f:
    df1 = pd.read_csv("/home/astegic/python_workspace/x/input.csv", skiprows= n, nrows=1000)
    x = df1.values.tolist()
    x1 = []
    for i in x:
        for c in i:
            x1.append(c)
    y = list(combinations(x1, 2))   

    xg = []
    for i in x1:
        xg.append(i.split('.', 1)[0])
    y1 = list(combinations(xg, 2))

    z = []
    matching_positions = 0
    for i in y1:
        for c1, c2 in zip(i[0], i[1]):
            if c1 == c2:
                matching_positions += 1
        z.append(matching_positions)        
    z.insert(0, 0)
    z1 = [x[1] - x[0] for x in foo(z)]
    dfz1 = pd.DataFrame(z1, columns=['number_of_letters_match']) 

    do1 = []
    do2 = []
    for i in y:
        if len(i[0].split('.', 1)[0]) >= len(i[1].split('.', 1)[0]):
            do1.append(i[0])
            do2.append(i[1])
        else:    
            do1.append(i[1])
            do2.append(i[0])

    d1 = []
    d2 = []
    for i in y1:
        if len(i[0]) >= len(i[1]):
            d1.append(i[0])
            d2.append(i[1])
        else:    
            d1.append(i[1])
            d2.append(i[0])
            
    dfo1 = pd.DataFrame(do1, columns=['domain1']) 
    dfo2 = pd.DataFrame(do2, columns=['domain2']) 
    df1 = pd.DataFrame(d1, columns=['domain1_temp']) 
    df2 = pd.DataFrame(d2, columns=['domain2_temp']) 
    df = pd.concat([dfo1, dfo2, df1, df2], axis=1)
    df['total_number_of_letters'] = df['domain1_temp'].str.len()
    df = pd.concat([df, dfz1], axis=1)
    df['Difference'] = df['total_number_of_letters'] - df['number_of_letters_match']
    df = pd.DataFrame(df, columns=['domain1', 'domain2', 'total_number_of_letters', 'number_of_letters_match', 'Difference'])
    df = df.loc[df['Difference'] == 1]
    df.to_csv("/home/astegic/python_workspace/x/output_file_" + str(o) + ".csv", index=False)

    n += 1000
    o += 1

files = glob.glob("/home/astegic/python_workspace/x/output_file_*.csv") 
data = pd.concat((pd.read_csv(f, header = 0, low_memory=False) for f in files))
data.to_csv("/home/astegic/python_workspace/x/output.csv", index=False)

fileList = glob.glob("/home/astegic/python_workspace/x/output_file_*.csv")
for filePath in fileList:
    os.remove(filePath)
```
